```python
def elevation(lat, lng):
    apikey = "AIzaSyCg-g87QRpnQjsmCgua3HFVxbmiCMBfN7I"
    url = "https://maps.googleapis.com/maps/api/elevation/json"
    request = urllib.request.urlopen(url+"?locations="+str(lat)+","+str(lng)+"&key="+"AIzaSyCg-g87QRpnQjsmCgua3HFVxbmiCMBfN7I")
    try:
        results = json.load(request).get('results')
        if 0 < len(results):
            elevation = results[0].get('elevation')
            # ELEVATION
            return elevation
        else:
            print('HTTP GET Request failed.')
    except ValueError:
        print ('JSON decode failed: '+str(request))
```

```python
query = '''
SELECT
  lng.id
  ,decrypted_lng
  , decrypted_lat
FROM `rgpkorea-datascience.dict.crypted_longitudes_20171100` lng
LEFT OUTER JOIN `rgpkorea-datascience.dict.crypted_latitudes_20171100` lat ON lng.id = lat.id
LIMIT 100
'''
project_id = 'rgpkorea-datascience'
```

```python
df = pd.read_gbq(
                query=query
                , project_id=project_id
                , dialect='standard'
)
```

```python
df.head()
```

```python
def elevation_df(df):
    lat, lng = df['decrypted_lat'], df['decrypted_lng']
    apikey = "AIzaSyCg-g87QRpnQjsmCgua3HFVxbmiCMBfN7I"
    url = "https://maps.googleapis.com/maps/api/elevation/json"
    request = urllib.request.urlopen(url+"?locations="+str(lat)+","+str(lng)+"&key="+"AIzaSyCg-g87QRpnQjsmCgua3HFVxbmiCMBfN7I")
    try:
        results = json.load(request).get('results')
        if 0 < len(results):
            elevation = results[0].get('elevation')
            # ELEVATION
            return elevation
        else:
            print('HTTP GET Request failed.')
    except ValueError:
        print ('JSON decode failed: '+str(request))
```

```python
df['ele'] = df.apply(elevation_df, axis=1)
```
